```python
import pandas as pd
```

```python
# From https://worldpopulationreview.com/country-rankings/murder-rate-by-country
murders = pd.read_csv("data.csv")
```

```python
murders
```

```python
# From https://wid.world/data/
inequality = pd.read_csv("WID_Data_16052023-163904.csv", sep=";", skiprows=[0])
```

```python
inequality
```

```python
inequality.drop(columns=["Percentile"], inplace=True)
```

```python
inequality.set_index("Year", inplace=True)
```

```python
# Rename columns
mapper = dict(map(lambda name: (name, name.split("\n")[0].split("_")[-1]), inequality.columns.values))
for before, after in mapper.items():
    print(repr(before), after)
```

```python
inequality.rename(columns=mapper, inplace=True)
```

```python
inequality
```

```python
latest_inequality = inequality.ffill().T[2021]
```

```python
latest_murders = murders.set_index("cca2")["latestRate"]
```

```python
i_frame = latest_inequality.to_frame(name="Top 1% Share of Income")
```

```python
m_frame = latest_murders.to_frame(name="Murder Rate (per 100k)")
```

```python
len(set(i_frame.index) & set(m_frame.index))
```

```python
combined = pd.concat([i_frame, m_frame], axis=1).dropna()
```

```python
len(combined)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=combined["Top 1% Share of Income"], y=combined["Murder Rate (per 100k)"])
```

```python
combined.index.name = "Country"
```

```python
combined.sort_index(inplace=True)
```

```python
# To export data
# combined.to_csv("income_inequality_v_murder_rate.csv")
```

```python
from scipy.stats import linregress
```

```python
res = linregress(combined["Top 1% Share of Income"], y=combined["Murder Rate (per 100k)"])
```

```python
res
```

```python
print(f"R^2: {res.rvalue**2:.4f}")
```

```python
from matplotlib import ticker
x = combined["Top 1% Share of Income"]
y = combined["Murder Rate (per 100k)"]
plt.plot(x, y, '.', label=f"n={len(combined)}")
plt.plot(x, res.intercept + res.slope * x, 'r', label=f"R^2: {res.rvalue**2:.4f}")
plt.legend()
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f"{x:.0%}"))
plt.xlabel("Top 1% Share of Income")
plt.ylabel("Murder Rate (per 100k)")
plt.show()
```

```python
# Re-produce logged r**2
import numpy as np
logged_res = linregress(combined["Top 1% Share of Income"], y=np.log(combined["Murder Rate (per 100k)"]))
```

```python
print(f"R**2: {logged_res.rvalue**2:.4f}")
```

```python
x = combined["Top 1% Share of Income"]
y = combined["Murder Rate (per 100k)"]

plt.plot(x, np.log(y), '.', label=f"n={len(combined)}")
plt.plot(x, logged_res.intercept + logged_res.slope * x, 'r', label=f"log-y R^2: {logged_res.rvalue**2:.4f}")
plt.legend()
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f"{x:.0%}"))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f"{np.e**x:.1f}"))
plt.xlabel("Top 1% Share of Income")
plt.ylabel("Murder Rate (per 100k) - log axis")
plt.show()
```

```python
# The inequality data has a number of data points that all equal 0.1962
combined[combined["Top 1% Share of Income"] == 0.1962]
```

```python
# Remove these and re-do the linear regression
altered = combined[combined["Top 1% Share of Income"] != 0.1962].copy()
altered_res = linregress(altered["Top 1% Share of Income"], y=altered["Murder Rate (per 100k)"])
x = altered["Top 1% Share of Income"]
y = altered["Murder Rate (per 100k)"]
plt.plot(x, y, '.', label=f"n={len(altered)}")
plt.plot(x, altered_res.intercept + altered_res.slope * x, 'r', label=f"R^2: {altered_res.rvalue**2:.4f}")
plt.legend()
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f"{x:.0%}"))
plt.xlabel("Top 1% Share of Income")
plt.ylabel("Murder Rate (per 100k)")
plt.show()
```
